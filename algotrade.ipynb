```python
def calc_indicator(ts):
    df = ts.copy()
    close = df['close']
    
    #
    df['SMA25'] = ta.sma(close, length=25) 
    df['SMA75'] = ta.sma(close,length=75)
    df['SMA200'] = ta.sma(close,length=200)
    #RSI
    df['RSI14'] = ta.rsi(close,length=14)

    #Bollinger Bands (21)
    bbands1 = ta.bbands(close, length=21, std=1)
    bbands2 = ta.bbands(close, length=21, std=2)
    bbands3 = ta.bbands(close, length=21, std=3)

    df = pd.concat([df, bbands1, bbands2, bbands3], axis=1)

    #MACD
    macd = ta.macd(close, fast=12, slow=26, signal=9)
    df_ = pd.concat([df, macd], axis=1)
    
    df = df.dropna()
    
    return df
```

```python
ts = calc_indicator(ts)
features = ts.columns.to_list()
```
